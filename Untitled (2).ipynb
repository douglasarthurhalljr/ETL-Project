```python
# Pandas
import pandas as pd

# SQL Alchemy
from sqlalchemy import create_engine
```

```python
# Create Engine for basketball data
#basketball_database_path = "..///basketball.sqlite"
engine = create_engine("sqlite:///basketball.sqlite")
conn = engine.connect()
```

```python
# Query All Records in the the Census Table
basketball_data = pd.read_sql("SELECT * FROM Draft", conn)
basketball_data
```

```python
draft_cols = ["yearDraft", "numberPickOverall", "namePlayer"]
draft_transformed= basketball_data[draft_cols].copy()

# Rename the column headers
draft_transformed = draft_transformed.rename(columns={"namePlayer": "full_name", "yearDraft": "draft_year",
                                                          "numberPickOverall": "pick_overall"})

# Clean the data by dropping duplicates and setting the index
#draft_transformed.drop_duplicates("full_name", inplace=True)
#draft_transformed.set_index("full_name", inplace=True)

draft_transformed.head()
```

```python
basketball_data = pd.read_sql("SELECT * FROM Player_Salary", conn)
basketball_data
```

```python
salary_cols = ["nameTeam", "namePlayer","value"]
salary_transformed= basketball_data[salary_cols].copy()

# Rename the column headers
salary_transformed = salary_transformed.rename(columns={"nameTeam": "team_name",
                                                          "namePlayer": "full_name",
                                                       "value": "salary"})
salary_transformed.set_index("full_name", inplace=True)
salary_transformed.head()
```
